```python
def toimage(sample, shape):
    sample = (sample+1)*128
    sample[sample<0] = 0
    sample[sample>255] = 255
    sample = sample.cpu().numpy().astype('uint8').reshape(len(sample), *shape)
    return sample
```

## Load data

```python
def part_wd(X,Xp,partition = 500,norm_var=False,eps=1e-4,max_iter=100):
    sinkhorn = SinkhornDistance(eps=eps, max_iter=max_iter)
    n_samples = X.shape[0]
    idx = 0
    avg_wd = 0
    while idx < n_samples-partition:
        
        avg_wd += partition * sinkhorn(X[idx:idx+partition],Xp[idx:idx+partition])
        idx += partition
    
    partition = n_samples - idx
    avg_wd += partition * sinkhorn(X[idx:idx+partition],Xp[idx:idx+partition])

    if norm_var:
        avg_wd = avg_wd/torch.var(X) 
    
    return avg_wd/n_samples
```

```python
verbose = True
shape = [28,28,1]
ndim = shape[0]*shape[1]*shape[2]
```

```python
X_train = torch.load('')
y_train = torch.load('')

X_cnn = torch.load('')
y_cnn = torch.load('')
X_eval = torch.load('')
y_eval = torch.load('')

X_test = np.concatenate((X_cnn, X_eval))
y_test = np.concatenate((y_cnn, y_eval))
```

## 1->0

```python
source_class = 0
target_class = 1

shape = [28,28,1]

data_train = torch.Tensor(X_train[y_train == source_class])
sample = torch.Tensor(X_train[y_train == target_class])

data_test = torch.Tensor(X_test[y_test == source_class])
sample_test = torch.Tensor(X_test[y_test == target_class])
sample_test_original = sample_test.clone()
data_test_fid = toimage(data_test, shape)
sample_test_original_fid = toimage(sample_test_original, shape)

nsample_A = len(data_train) 
nsample = 6*len(data_train)
```

```python
wd = part_wd(data_test,sample_test)
print('data')
plot_images(1-data_test[:8, :])
plt.show()
print('sample')
plot_images(1-sample_test[:8, :])        
plt.show()
print(wd)
```

```python
model = SINF(ndim=ndim).requires_grad_(False)
```

```python
Niter = 50
update_iteration = 5
K = 56
nlayer = 0
batchsize=None
FID = []
wd_list = []
wd = part_wd(data_test,sample_test)
print('data')
plot_images(1-data_test[:8, :])
plt.show()
print('sample')
plot_images(1-sample_test[:8, :])        
plt.show()
print(wd) 
wd_list.append([0,wd,wd]) #[layer,wd_for,wd_inv]

timer = time.time()
total_time=0

for i in range(Niter):
    nlayer += 1
    if nlayer <= len(model.layer):
        continue
    model, sample, sample_test = add_one_layer_inverse(model, data_train, sample, K, nsample_A, nsample, layer_type='regular', 
                                                        batchsize=batchsize, sample_test=sample_test, put_data_on_disk=args.put_data_on_disk)
    if len(model.layer) % update_iteration == 0:
        print()
        print('Finished %d iterations' % len(model.layer), 'Total Time:', time.time()- timer + total_time)
        total_time = time.time()-timer + total_time
        
        print()

        wd = part_wd(data_test,sample_test)
        
        inv_sample_test,_ = model(data_test)
        wd_inv = part_wd(sample_test_original,inv_sample_test)
        print('data')
        plot_images(1-data_test[:8, :])
        plt.show()
        print('sample')
        plot_images(1-sample_test[:8, :])        
        plt.show()
        print('inverse sample')
        plot_images(1-inv_sample_test[:8, :])        
        plt.show()
        print('wd=',wd)
        print('inv_wd=',wd_inv)
        wd_list.append([i,wd,wd_inv])
        # torch.save(model, args.save)

        sample_test1 = toimage(sample_test, shape)
        fid = evaluate_fid_score(sample_test1.astype(np.float32)/255., data_test_fid.astype(np.float32)/255.)
        inv_sample_test1 = toimage(inv_sample_test, shape)
        fid_inv = evaluate_fid_score(inv_sample_test1.astype(np.float32)/255., data_test_fid.astype(np.float32)/255.)
        FID.append([fid,fid_inv])
        print('Current FID score:', fid)
        print('Current inverse FID score:', fid_inv)
        #del sample_test1
        timer =time.time()
```

```python
plot_images(1-sample_test_original[:8, :])
b,_ = model.inverse(sample_test_original)
plot_images(1-b[:8, ]) 
```

```python
plot_images(1-data_test[:8, :])
b,_ = model(data_test)
plot_images(1-b[:8, :]) 
```

```python
wd_tensor10 = torch.FloatTensor(wd_list)
fid_tensor10 = torch.FloatTensor(FID)
min_wd = torch.argmin(wd_tensor10[:,1])
min_fid = torch.argmin(fid_tensor10[:,0])
```

```python
import matplotlib.pyplot as plt
plt.plot(wd_tensor10[:,0],wd_tensor10[:,1],label='1->0')
plt.plot(wd_tensor10[:,0],wd_tensor10[:,2],label='0->1')
plt.xlabel('layer')
plt.ylabel('test WD')
plt.title('1->0')
plt.legend()
plt.show()
print(f'after layer{wd_tensor10[:,0][min_wd]}, wd ={wd_tensor10[:,1][min_wd]}')
```

```python
plt.plot(wd_tensor10[1:,0],fid_tensor10[:,0],label='1->0')
plt.plot(wd_tensor10[1:,0],fid_tensor10[:,1],label='0->1')
plt.xlabel('layer')
plt.ylabel('FID')
plt.title('1->0')
plt.show()
print(f'after layer{wd_tensor10[:,0][min_fid+1]}, FID ={fid_tensor10[min_fid,0]}')
print(f'after layer{wd_tensor10[:,0][min_wd]}, FID ={fid_tensor10[min_wd-1,0]}')
```

```python
avg_wd_tensor = (wd_tensor10[:,1]+wd_tensor10[:,2])/2
avg_fid_tensor = (fid_tensor10[:,0]+fid_tensor10[:,1])/2
min_wd = torch.argmin(avg_wd_tensor)
min_fid = torch.argmin(avg_fid_tensor)


print(f'after layer{wd_tensor10[min_wd,0]}, WD ={avg_wd_tensor[:][min_wd]}')
print(f'after layer{wd_tensor10[min_fid+1,0]}, FID ={avg_fid_tensor[min_fid]} (based on FID)')
print(f'after layer{wd_tensor10[min_wd,0]}, FID ={avg_fid_tensor[min_wd-1]}')

```
