# MNIST (019)

```python
root_dir = '/scratch/gilbreth/zhou1059/IAF/saved_images/mnist019'
```

```python
root_dir
```

```python
import os 

exp_list = ['exp3','exp4','exp5']
class_list = [0,1,9]
key_list = ['GB','NB','GBNB','mSWD-NB','DD']

for exp_idx in exp_list:
    exp_dir = os.path.join(root_dir, exp_idx)
    
    for key in key_list:
        print('##################################')
        print(f'############ {key} ##############')
        print('##################################')
        for c in class_list:
            for m in class_list:
                print(f'############ {c} to {m} ##############')
                real_path = os.path.join(exp_dir,'real')
                real_path = os.path.join(real_path,f'{c}')

                flip_path = os.path.join(exp_dir,f'{key}')
                flip_path = os.path.join(flip_path,f'{c}')
                flip_path = os.path.join(flip_path,f'{m}')
                !python -m pytorch_fid {real_path} {flip_path} --device cuda:0
                #!python -m pytorch_fid {real_path} {flip_path} 
```
